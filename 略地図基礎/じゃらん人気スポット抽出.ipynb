```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

# Chromeオプション設定
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

service = Service("/usr/bin/chromedriver")
driver = webdriver.Chrome(service=service, options=options)

# ページにアクセス
url = "https://www.jalan.net/kankou/140000/140200/"
driver.get(url)

# 各スポットの親ブロック（施設全体）を取得
spots = driver.find_elements(By.CLASS_NAME, "item-info")  # 各スポットの親要素（Jalan構造によく使われる）

for spot in spots:
    try:
        # スポット名リンクを取得（myFormが含まれているもの）
        name_link = spot.find_element(By.XPATH, ".//a[contains(@onclick, 'myForm') and contains(@href, 'kankou/spt_')]")
        spot_name = name_link.text.strip()

        # 口コミリンクを取得（myFormかつkankou/spt_guide）
        review_elem = spot.find_element(By.CLASS_NAME, "reviewCount")
        review_link = review_elem.find_element(By.TAG_NAME, 'a')
        href = review_link.get_attribute("href")
        onclick = review_link.get_attribute("onclick")

        # 条件：hrefにkankou/spt_guide、onclickにmyForm
        if "kankou/spt" in href and "myForm" in onclick:
            review_text = review_link.text.strip()
            print(f"🏷️ {spot_name} - 📝 {review_text}")

    except Exception:
        continue

driver.quit()
```

```python
import csv
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time

# Chromeオプション設定
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
service = Service("/usr/bin/chromedriver")
driver = webdriver.Chrome(service=service, options=options)

# 出力CSVファイル名
csv_filename = "jalan_spots.csv"

# ヘッダーを書き込み（最初だけ）
with open(csv_filename, mode='w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["スポット名", "口コミ件数"])

# 初期ページ
url = "https://www.jalan.net/kankou/140000/140200/"
driver.get(url)
time.sleep(2)

page_count = 1

while True:
    print(f"\n📄 ページ {page_count} のデータ取得中...")

    spots = driver.find_elements(By.CLASS_NAME, "item-info")

    for spot in spots:
        try:
            name_link = spot.find_element(By.XPATH, ".//a[contains(@onclick, 'myForm') and contains(@href, 'kankou/spt_')]")
            spot_name = name_link.text.strip()

            review_elem = spot.find_element(By.CLASS_NAME, "reviewCount")
            review_link = review_elem.find_element(By.TAG_NAME, 'a')
            href = review_link.get_attribute("href")
            onclick = review_link.get_attribute("onclick")

            if "kankou/spt" in href and "myForm" in onclick:
                review_text = review_link.text.strip()
                print(f"🏷️ {spot_name} - 📝 {review_text}")

                # CSVに追記
                with open(csv_filename, mode='a', encoding='utf-8', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow([spot_name, review_text])

        except Exception:
            continue

    # 次のページへ進む
    try:
        next_link = driver.find_element(By.CLASS_NAME, "txtLinks.next")
        next_link.click()
        time.sleep(2)
        page_count += 1
    except NoSuchElementException:
        print("\n✅ すべてのページの取得が完了しました。")
        break

driver.quit()
```

```python
import pandas as pd

# 読み込むファイル名（元のCSV）
filename = 'jalan_spots.csv'

# CSVを読み込む
df = pd.read_csv(filename)

# 口コミ件数から数値だけを抽出し、「口コミ数」列として追加
df['口コミ数'] = df['口コミ件数'].str.replace('口コミ', '', regex=False)\
                                   .str.replace('件', '', regex=False)\
                                   .str.replace(',', '', regex=False)\
                                   .astype(int)

# 元のファイル名で上書き保存（index=Falseでインデックス列なし）
df.to_csv(filename, index=False)

print(f"{filename} に上書き保存しました。")
```
